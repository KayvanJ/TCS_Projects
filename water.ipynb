```python
import pandas as pd

df = pd.read_csv("water_insecurity_2023.csv", encoding = 'latin-1')

df.head()
```

```python
state = []

county = []
for county_state in df.name:
    state.append(county_state.split(', ')[1])
    
    county.append(county_state.split(', ')[0])

df['state'] = state

df['county'] = county
```

```python
import matplotlib.pyplot as plt
plumb_data = df.groupby('state')['plumbing'].sum().sort_values(ascending=False)


plt.figure(figsize = (15,5))
plt.bar(plumb_data.index, plumb_data)

plt.xticks(rotation = 90)
plt.show()
```
